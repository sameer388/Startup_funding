```python
import csv
import numpy as np
import matplotlib.pyplot as plt
file_obj = open("D:/Data Science/14. Project-Case Study(Part-1)/startup_funding.csv",encoding = 'utf8')
file_data = csv.DictReader(file_obj,skipinitialspace = True)
file_list = list(file_data)
```

```python
dic = {}
for row in file_list:
    if row['IndustryVertical'] == 'eCommerce':
        row['IndustryVertical'] = 'Ecommerce'
    if row['IndustryVertical'] == 'ECommerce':
        row['IndustryVertical'] = 'Ecommerce'
    if row['IndustryVertical'] == 'ecommerce':
        row['IndustryVertical'] = 'Ecommerce'
    if row['AmountInUSD'] != '' and row['IndustryVertical'] != '':
        x = row['AmountInUSD'].replace(',','')
        if row['IndustryVertical'] in dic:
            dic[row['IndustryVertical']] += int(float(x))
        else:
            dic[row['IndustryVertical']] = int(float(x))
Type = []
amount = []
for i in dic.keys():
    Type.append(i)
    amount.append(dic[i])
np_type = np.array(Type)
np_amount = np.array(amount)
np_type = np_type[np.argsort(np_amount)]
np_amount = np.sort(np_amount)
l = len(np_type)
np_type = np_type[l-1:l-6:-1]
np_amount = np_amount[l-1:l-6:-1]
total = sum(np_amount)
for i in range(len(np_type)):
    print(np_type[i],round((np_amount[i]/total)*100,2))
plt.bar(np_type,np_amount,width = 0.5)
plt.xticks(rotation = 40)
plt.xlabel('Sector',size  = 20)
plt.show()
```
