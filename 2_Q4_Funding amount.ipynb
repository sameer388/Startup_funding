```python
import csv
import matplotlib.pyplot as plt
file_obj = open("D:/Data Science/14. Project-Case Study(Part-1)/startup_funding.csv",encoding = 'utf8')
file_data = csv.DictReader(file_obj,skipinitialspace = True)
file_list = list(file_data)
```

```python
import numpy as np
dic = {}
for row in file_list:
    #if row['AmountInUSD'] == '':
        #row['AmountInUSD'] == '0.0'
    #else:
        #row['AmountInUSD'] = row['AmountInUSD'].replace(',','')
    if row['CityLocation'] == 'bangalore':
        row['CityLocation'] = 'Bangalore'
    if row['CityLocation'] == 'Delhi':
        row['CityLocation'] = 'New Delhi'
    row['CityLocation'] = row['CityLocation'].split('/')[0].strip()
    if row['AmountInUSD'] != '' and row['CityLocation'] != '':
        if row['CityLocation'] in dic:
            dic[row['CityLocation']] += int(float(row['AmountInUSD']))
        else:
            dic[row['CityLocation']] = int(float(row['AmountInUSD']))
location = []
amount = []
for i in dic:
    location.append(i)
    amount.append(dic[i])
np_location = np.array(location)
np_amount = np.array(amount)
np_location = np_location[np.argsort(np_amount)]
np_amount = np.sort(np_amount)
l = len(np_location)
np_location = np_location[l-1:l-11:-1]
np_amount = np_amount[l-1:l-11:-1]
np_full = np.full(10,0.6)
plt.pie(np_amount,labels = np_location,explode = np_full,radius = 2,autopct = "%0.2lf%%")
plt.show()
total = sum(np_amount)
percentage = []
for i in np_amount:
    x = (i/total)*100
    percentage.append(round(x,2))
for i in range(10):
    print(np_location[i],percentage[i])
```
