```python
import csv
import matplotlib.pyplot as plt
file_obj = open("D:/Data Science/14. Project-Case Study(Part-1)/startup_funding.csv",encoding = 'utf8')
file_data = csv.DictReader(file_obj,skipinitialspace = True)
file_list = list(file_data)
file_list
```

Ques1-Find the trend of investing in india over the years.Draw an line graph of investing vs year in india.


```python
year = []
for row in file_list:
    year.append(row['Date'][len(row['Date'])-4:])
dic = {}
for i in year:
    if i in dic:
        dic[i] += 1
    else:
        dic[i] = 1
Year = []
Investment = []
for i in dic:
    Year.append(i)
    Investment.append(dic[i])
Year.reverse()
Investment.reverse()
plt.plot(Year,Investment,linewidth = 4)
plt.xlabel("Year",size = 20)
plt.ylabel("Investment",size = 20)
plt.xticks(size = 20)
plt.yticks(size = 20)
plt.show()
for i in range(len(Year)):
    print(Year[i],Investment[i])
```

Ques2- Find the top 10 indian companies which have most no of startups.
Print a Pie charts and visualize it and print the city and no startups in decreasing order


```python
location = []
for row in file_list:
    location.append(row['CityLocation'])
import numpy as np
np_location = np.array(location)
np_location = np_location[np_location != '']
for i in range(len(np_location)): ## for changing the value at delhi and bangalore
    if np_location[i] == 'bangalore':
        np_location[i] = 'Bangalore'
    if np_location[i] == 'Delhi':
        np_location[i] = 'New Delhi'
for i in range(len(np_location)):
    np_location[i] = np_location[i].split('/')[0].strip() ## for knowing the things other than india
dic = {}
for i in np_location:
    if i in dic:
        dic[i] += 1
    else:
        dic[i] = 1
city = []
Number = []
for i in dic:
    city.append(i)
    Number.append(dic[i])
np_full = np.full(10,0.5)
np_city = np.array(city)
np_Number = np.array(Number)
np_city = np_city[np.argsort(np_Number)]
np_Number = np.sort(np_Number)
np_city = np_city[len(np_city)-1:len(np_city)-11:-1]
np_Number = np_Number[len(np_Number)-1:len(np_Number)-11:-1]
plt.pie(np_Number,labels = np_city,radius = 2,autopct = "%0.2lf%%",explode =np_full,textprops = {'fontsize' : 16} )
plt.show()
for i in range(len(np_city)):
    print(np_city[i],np_Number[i])
```
